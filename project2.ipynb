## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = 'https://en.wikipedia.org/wiki/Lamborghini'
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")

for data in soup(['style', 'script']):
  data.decompose()

print(' '.join(soup.stripped_strings))
```

```python
#download an image with beautifulsoup and save it in this repository
image_url = 'https://upload.wikimedia.org/wikipedia/commons/3/3b/Lamborghini%2C_Sant%27Agata_Bolognese%2C_Bologna.jpg'
image_response = requests.get(image_url)
with open('lamborghini.jpg', 'wb') as file:
    file.write(image_response.content)
```

```python
links = soup.find_all('a', href=True)

# Loop through the first 10 links and print the first 100 characters
for i in range(11):
    link = links[i].get('href')
    print(f"Link {i+1}: {link[:100]}")
```

```python
#download the movie posters for 10 of these movies and save them to this repository
import os

poster_folder = 'posters'
os.makedirs(poster_folder, exist_ok=True)

for movie in data['results']:
    poster_path = movie.get('poster_path')
    poster_url = f"https://image.tmdb.org/t/p/w500{poster_path}"

    poster_name = movie.get('title').replace(" ", "_") + '_poster.jpg'
    poster_path_local = os.path.join(poster_folder, poster_name)

    response = requests.get(poster_url)

    with open(poster_path_local, "wb") as file:
        file.write(response.content)
```
